```python
import pandas as pd
import os
import numpy as np

df = pd.read_csv('../documents/aliexpress laptops.csv')
```

```python
# 1) Define your bins and labels
bins = [0, 1000, 3000, 5000, 7000,  10000, 20000, np.inf]            
labels = ["0–1000", "1001–3000", "3001-5000", "5001-7000",
          "7001-10000", "10001-20000", "20000+"]

# 2) Create a new column categorizing each price into a bin
df["price_range"] = pd.cut(df["price"], bins=bins, labels=labels, right=True)
grouped = df.groupby("price_range",  observed=True).count()
```

```python
ax = grouped["name"].plot()
fig = ax.get_figure()
```
